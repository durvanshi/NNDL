```python
import numpy as np
vector1= np.array([10,20,30])
vector2= np.array([5,10,15])
print("Vector1 = ", vector1 )
print("Vector2 = ", vector2)
result = vector1 + vector2
print("Vector addition = " , result)
```

```python
import numpy as np
m1 = np.array([[2,1,3],[4,0,2]])
m2 = np.array([[5,2],[1,4],[3,7]])
print("Matrix 1 = ", m1)
print("Matrix 2 = ", m2)
res = np.matmul(m1,m2)
print("Matrix multiplication = \n",res)
```
